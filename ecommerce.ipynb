```python
import pandas as pd

df_cust = pd.read_csv('/Users/shraddhagupta/Downloads/archive/olist_customers_dataset.csv')
df_geo = pd.read_csv('/Users/shraddhagupta/Downloads/archive/olist_geolocation_dataset.csv')
df_oitems = pd.read_csv('/Users/shraddhagupta/Downloads/archive/olist_order_items_dataset.csv')
df_opay = pd.read_csv('/Users/shraddhagupta/Downloads/archive/olist_order_payments_dataset.csv')
df_oreview = pd.read_csv('/Users/shraddhagupta/Downloads/archive/olist_order_reviews_dataset.csv')
df_order = pd.read_csv('/Users/shraddhagupta/Downloads/archive/olist_orders_dataset.csv')
df_prod = pd.read_csv('/Users/shraddhagupta/Downloads/archive/olist_products_dataset.csv')
df_sell = pd.read_csv('/Users/shraddhagupta/Downloads/archive/olist_sellers_dataset.csv')
df_pcat = pd.read_csv('/Users/shraddhagupta/Downloads/archive/product_category_name_translation.csv')

df_deal = pd.read_csv('/Users/shraddhagupta/Downloads/olist_closed_deals_dataset.csv')

```

```python
df_seller = df_deal.merge(df_sell, how='inner', on='seller_id')

df_seller = df_seller[['seller_id', 'seller_state', 'seller_city', 'business_segment', 'lead_type', 'business_type']]

df_seller.info()
```

```python
df_final1 = df_seller.merge(df_oitems, how='inner', on='seller_id')

df_final1.info()
```

```python
df_product = df_prod.merge(df_pcat, how='inner', on='product_category_name')

df_product = df_product[['product_id', 'product_category_name_english']]

df_product.info()
```

```python
df_final2 = df_final1.merge(df_product, how='inner', on='product_id')

df_final2.info()
```

```python
df_final3 = df_final2.merge(df_order, how='inner', on='order_id')

df_final3 = df_final3[['seller_id', 'seller_state', 'seller_city', 'business_segment',
       'lead_type', 'business_type', 'order_id', 'order_item_id', 'product_id',
       'shipping_limit_date', 'price', 'freight_value',
       'product_category_name_english', 'customer_id', 'order_delivered_customer_date', 'order_estimated_delivery_date']]

df_final3.info()
```

```python
df_final4 = df_final3.merge(df_oreview, how='inner', on='order_id')

df_final4 = df_final4[['seller_id', 'seller_state', 'seller_city', 'business_segment',
       'lead_type', 'business_type', 'order_id', 'order_item_id', 'product_id',
       'shipping_limit_date', 'price', 'freight_value',
       'product_category_name_english', 'customer_id', 'order_delivered_customer_date', 'order_estimated_delivery_date', 'review_score']]

df_final4.info()
```

```python
df_final5 = df_final4.merge(df_opay, how='inner', on='order_id')

df_final5 = df_final5[['seller_id', 'seller_state', 'seller_city', 'business_segment',
       'lead_type', 'business_type', 'order_id', 'order_item_id', 'product_id',
       'shipping_limit_date', 'price', 'freight_value',
       'product_category_name_english', 'customer_id', 'order_delivered_customer_date', 'order_estimated_delivery_date', 'review_score',
       'payment_type', 'payment_value']]

df_final5.info()
```

```python
df_final = df_final5.merge(df_cust, how='inner', on='customer_id')

df_final = df_final[['seller_id', 'seller_state', 'seller_city', 'business_segment',
       'lead_type', 'business_type', 'order_id', 'order_item_id', 'product_id',
       'shipping_limit_date', 'price', 'freight_value',
       'product_category_name_english', 'customer_id', 'order_delivered_customer_date', 'order_estimated_delivery_date', 'review_score',
       'payment_type', 'payment_value', 'customer_city', 'customer_state']]

df_final.info()
```

```python
df_final.to_csv('ecomm.csv', index=False) 
```
